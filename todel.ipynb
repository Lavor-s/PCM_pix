```python
import gdspy
import numpy as np

gdspy.current_library = gdspy.GdsLibrary()
lib = gdspy.GdsLibrary()
cell = lib.new_cell("CELL_0")
layer = {"layer": 0, "datatype": 0}
multipl = 1e6
tolerance = 0.001
x_center = 0
y_center = 0
rad = 500e-9
b_in = 50e-9

cell.add(
                    gdspy.Round(
                        (x_center * multipl, y_center * multipl),
                        rad * multipl,
                        inner_radius=b_in * multipl,
                        initial_angle=-np.pi,
                        final_angle=np.pi,
                        tolerance=tolerance,
                        **layer,
                    )
                )

lib.write_gds('test.gds')
```
